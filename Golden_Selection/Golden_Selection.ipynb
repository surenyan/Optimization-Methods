```python
import numpy as np
import plotly.graph_objects as go

def f(x):
    return x**3 - 6*x**2 + 9*x + 1

a, b = 1, 5
iterations = []
```

```python
while abs(b - a) > 0.5:
    x1, x2 = a + 0.3819 * (b - a), a + 0.6180 * (b - a)
    iterations.append((a, b, x1, x2, f(x1), f(x2)))
    if f(x1) < f(x2):
        b = x2
    else:
        a = x1

print(f"xmin = {(a+b)/2:.4f} | f(xmin) = {f((a+b)/2):.4f}")
```

```python
#plot
x_full = np.linspace(0.5, 5.5, 400)
y_full = f(x_full)

for i, (a_i, b_i, x1, x2, f_x1, f_x2) in enumerate(iterations):
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=x_full, y=y_full, mode='lines', name='f(x)', line=dict(color='blue')))
    fig.add_trace(go.Scatter(x=[x1], y=[f_x1], mode='markers', name='x1',
        marker=dict(size=12, color='red'), hovertemplate=f'x1={x1:.4f}<br>f(x1)={f_x1:.4f}<extra></extra>'))
    fig.add_trace(go.Scatter(x=[x2], y=[f_x2], mode='markers', name='x2',
        marker=dict(size=12, color='orange'), hovertemplate=f'x2={x2:.4f}<br>f(x2)={f_x2:.4f}<extra></extra>'))
    fig.add_shape(type="line", x0=a_i, y0=min(y_full), x1=a_i, y1=max(y_full), line=dict(color="green", width=2, dash="dot"))
    fig.add_shape(type="line", x0=b_i, y0=min(y_full), x1=b_i, y1=max(y_full), line=dict(color="green", width=2, dash="dot"))
    fig.update_layout(title=f"Iteration {i+1} | [{a_i:.4f}, {b_i:.4f}]",
        xaxis_title="x", yaxis_title="f(x)", xaxis=dict(range=[0.5, 5.5]), hovermode='closest')
    fig.show()
```
